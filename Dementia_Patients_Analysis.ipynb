# 1. Data Loading and Preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings


warnings.filterwarnings("ignore", category=FutureWarning, module="seaborn")
# Load the dataset
df = pd.read_csv("C:\\Users\\Pc\\Downloads\\dementia_patients_health_data.csv")
df
```

```python
missing_values = df.isna().sum()
missing_values
```

```python
df['Prescription'].fillna('None', inplace = True)
df['Dosage in mg'].fillna(0, inplace = True)
```

```python
df['Prescription'].isna().sum()
```

```python
df['Dosage in mg'].isna().sum()
```

```python
# Convert categorical columns to category dtype
categorical_columns = ['Education_Level', 'Dominant_Hand', 'Gender', 'Family_History', 'Smoking_Status',
                       'APOE_ε4', 'Physical_Activity', 'Depression_Status', 'Medication_History', 
                       'Nutrition_Diet', 'Sleep_Quality', 'Chronic_Health_Conditions', 'Prescription']

for col in categorical_columns:
    df[col] = df[col].astype('category')
```

```python
# Verify the preprocessing
print(df.info())
```

## 2. Exploratory Data Analysis (EDA)

#### a) Distribution of Age and Gender

```python
# Plot Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True, bins=20)
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot Gender distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

#### b) Prescription Patterns

```python
# Top 4 prescribed medications
top_prescriptions = df['Prescription'].value_counts().head(4)

# Plot the top 4 prescribed medications
plt.figure(figsize=(10, 6))
sns.barplot(x=top_prescriptions.values, y=top_prescriptions.index, data=df)
plt.title('Top 4 Prescribed Medications')
plt.xlabel('Number of Prescriptions')
plt.ylabel('Medication Name')
plt.show()
```

## 3. Visualization and Insights

#### a) Heart Rate and Blood Oxygen Level Analysis

```python
# Scatter plot of Heart Rate vs Blood Oxygen Level colored by Dementia status
plt.figure(figsize=(10, 6))
sns.scatterplot(x='HeartRate', y='BloodOxygenLevel', hue='Dementia', data=df)
plt.title('Heart Rate vs Blood Oxygen Level (Colored by Dementia Status)')
plt.xlabel('Heart Rate (bpm)')
plt.ylabel('Blood Oxygen Level (%)')
plt.show()
```

#### b) Impact of Smoking and Physical Activity on Dementia

##### Countplot for Smoking Status vs Dementia

```python
# Grouping and counting
counts = df.groupby(['Smoking_Status', 'Dementia']).size().unstack()

# Replacing NaN with 0 if there are any missing combinations
counts = counts.fillna(0)

counts.plot(kind='bar', figsize=(10, 6))

plt.title('Smoking Status vs Dementia')
plt.xlabel('Smoking Status')
plt.ylabel('Count')
plt.xticks(rotation=0) 
plt.legend(title='Dementia')
plt.show()
```

## 4. Correlation and Statistical Analysis

##### a) Correlation Heatmap

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()
# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

## 5. Conclusion and Summary

 The dataset provided valuable insights into the health profiles of dementia patients.
- **Age and Gender:** The age distribution showed most patients were elderly, and there was a relatively balanced gender distribution.
- **Prescription Patterns:** Commonly prescribed medications included Donepezil, Memantine, and Galantamine, which are often used to manage dementia symptoms.
- **Lifestyle Factors:** Smoking and physical inactivity appeared to be more common among patients with dementia, suggesting a potential link.
- **Heart Rate and Blood Oxygen Levels:** No strong correlation with dementia was found, but further analysis could be warranted.
- **Overall Health Correlation:** The correlation heatmap revealed some interesting relationships between variables, such as a strong negative correlation between cognitive test scores and dementia.

These insights could help in better understanding the factors contributing to dementia and potentially guiding interventions.
